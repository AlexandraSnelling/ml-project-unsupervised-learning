# Part II - KMeans Clustering

The objective of the analysis is to group similar products together into clusters based on their attributes such as fresh, milk, grocery, frozen, detergents_paper, and delicatessen. To perform the k-means clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters, initialize the centroids, assign data points to clusters, update the centroids, and repeat until convergence.

```python
X = df_numeric
```

```python
def plot_distortion(X,max_clusters = 20):
    distortions = []
    for i in range(1, max_clusters +1):
        km = KMeans(n_clusters=i,
                    init='k-means++',
                    n_init=50,
                    random_state=0)
        km.fit(X)
        distortions.append(km.inertia_)
    
    plt.plot(range(1,max_clusters +1), distortions, marker='o')
    plt.xlabel('Number of clusters')
    plt.ylabel('Distortion')
    plt.show() 
```

```python
# elbow rule
plot_distortion(X,max_clusters=15)
```

Based on the elbow curve, it appears that between 2-4 clusters will be optimal.

```python
for i in range(2,16):  
    # Fit the model
    km = KMeans(n_clusters=i, # how many clusters we expected 
                n_init=50, # how many initial runs
                random_state=0)

    # fit and predict
    km_clust = km.fit_predict(X)

    print("silhouette score for "+ str(i) + " clusters: " + str(silhouette_score(X, km_clust)))
```

The best (highest) silhouette score is obtained with only 2 clusters.

```python
for n_init_test in [10,50,100]:  
    # Fit the model
    km = KMeans(n_clusters=2, # how many clusters we expected 
                n_init=i, # how many initial runs
                random_state=0)

    # fit and predict
    km_clust = km.fit_predict(X)
    
    print(km.labels_[:37])

    print("silhouette score for "+ str(i) + " clusters: " + str(silhouette_score(X, km_clust)))
```

# Part III - Hierarchical Clustering 

Hierarchical clustering is a popular unsupervised machine learning algorithm that is used to identify patterns and group similar data points together in a hierarchy. The algorithm works by iteratively merging or splitting clusters based on a similarity measure until a dendrogram is formed.

To perform hierarchical clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters using techniques such as dendrogram.

```python
X = df_numeric
```

```python
print('simple linkage:')
for n_clusters in range(2,6): 
    # Fit the model
    ac = AgglomerativeClustering(affinity='euclidean',
                             linkage = 'single',
                             n_clusters = n_clusters)
    # fit and predict
    hc_clust = ac.fit_predict(X)
    #print result
    print("silhouette score for " + str(n_clusters) + " clusters: " + str(silhouette_score(X, hc_clust)))

print('complete linkage:')
for n_clusters in range(2,6): 
    # Fit the model
    ac = AgglomerativeClustering(affinity='euclidean',
                             linkage = 'complete',
                             n_clusters = n_clusters)
    # fit and predict
    hc_clust = ac.fit_predict(X)
    #print result
    print("silhouette score for " + str(n_clusters) + " clusters: " + str(silhouette_score(X, hc_clust)))

print('average linkage:')
for n_clusters in range(2,6): 
    # Fit the model
    ac = AgglomerativeClustering(affinity='euclidean',
                             linkage = 'average',
                             n_clusters = n_clusters)
    # fit and predict
    hc_clust = ac.fit_predict(X)
    #print result
    print("silhouette score for " + str(n_clusters) + " clusters: " + str(silhouette_score(X, hc_clust)))

print('wards linkage:')
for n_clusters in range(2,6): 
    # Fit the model
    ac = AgglomerativeClustering(affinity='euclidean',
                             linkage = 'ward',
                             n_clusters = n_clusters)
    # fit and predict
    hc_clust = ac.fit_predict(X)
    #print result
    print("silhouette score for " + str(n_clusters) + " clusters: " + str(silhouette_score(X, hc_clust)))
```

Again we see the best silhouette score for 2 clusters in both simple and average linkage. 

```python
ac = AgglomerativeClustering(affinity='euclidean',
                             linkage = 'single',
                             n_clusters = 2)
# fit and predict
hc_clust = ac.fit_predict(X)
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

# define plot_dendrogram function
def plot_dendrogram(X, method='single', figsize=(20, 6)):
    plt.figure(figsize=figsize)  # Adjust the width and height of the figure
    dendrogram = sch.dendrogram(sch.linkage(X, method=method), color_threshold=46500)
    plt.title("Dendrogram")
    plt.ylabel("Euclidean distances")
    plt.xlabel('Points')
    plt.show()
```

```python
plot_dendrogram(X)
```

```python
ac = AgglomerativeClustering(affinity='euclidean',
                             linkage = 'ward',
                             n_clusters = 6)
# fit and predict
hc_clust = ac.fit_predict(X)
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

# define plot_dendrogram function
def plot_dendrogram(X, method='ward', figsize=(20, 6)):
    plt.figure(figsize=figsize)  # Adjust the width and height of the figure
    dendrogram = sch.dendrogram(sch.linkage(X, method=method), color_threshold=115000)
    plt.title("Dendrogram")
    plt.ylabel("Euclidean distances")
    plt.xlabel('Points')
    plt.show()
```

# Part IV - PCA

In this section you are going to perform principal component analysis (PCA) to draw conclusions about the underlying structure of the wholesale customer data. Since using PCA on a dataset calculates the dimensions which best maximize variance, we will find which compound combinations of features best describe customers.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=None)

df_pca = pca.fit(X)
```

```python
for i in range(0,6):
    print("Component " + str(i+1) + " : " + df_numeric.columns[i])

plt.figure(figsize=(10,6))
plt.scatter(x=[i+1 for i in range(len(df_pca.explained_variance_ratio_))],
            y=df_pca.explained_variance_ratio_,
           s=200, alpha=0.75,c='orange',edgecolor='k')
plt.grid(True)
plt.title("Explained variance ratio of the \nfitted principal component vector\n",fontsize=25)
plt.xlabel("Principal components",fontsize=15)
plt.xticks([i+1 for i in range(len(df_pca.explained_variance_ratio_))],fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel("Explained variance ratio",fontsize=15)
plt.show()
```

We can see that the Fresh and Milk features of the dataset are the Principal Components.
